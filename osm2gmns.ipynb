**Step3: extract network by given link_type**

| param | value |descripion|
|--|--|--|
| link_type | primary, trunk ,residential, service,... | specifying the link type  to extract |


```python
net = og.getNetFromOSMFile(osm_filename='/content/osm_datasets/tempe_asu.osm',network_type=('auto'),POIs=True,link_type='primary')
og.outputNetToCSV(net)
net=pg.readNetwork()
pg.showNetByAllMode(net)
```

**Step4: extract network by given POI_sampling_ratio**

| param | value |descripion|
|--|--|--|
| POI_sampling_ratio | （0, 1]） | specifying the POI ratio to extract |

```python
net = og.getNetFromOSMFile(osm_filename='/content/osm_datasets/tempe_asu.osm',network_type=('auto'),POIs=True,POI_sampling_ratio=0.6)
og.outputNetToCSV(net)
net=pg.readNetwork()
pg.showNetByAllMode(net)
```

**Step5: extract network with min_nodes>1**

| param | value |descripion|
|--|--|--|
| min_nodes | 0, 1, 2,... | specifying the minimum number of nodes in a link that can be retained |

```python
net = og.getNetFromOSMFile(osm_filename='/content/osm_datasets/tempe_asu.osm',network_type=('auto'),POIs=True,min_nodes=10)
og.outputNetToCSV(net)
net=pg.readNetwork()
pg.showNetByAllMode(net)
```

**Step6 :extract network with given coordinate range**

| param | value |descripion|
|--|--|--|
| bbox | (minlat, minlon, maxlat, maxlon) | extract network with given coordinate range |

```python
box=(33.42,-111.94,33.43,-111.92)
net = og.getNetFromOSMFile(osm_filename='/content/osm_datasets/tempe_asu.osm',network_type=('auto'),POIs=True,bbox=box)
og.outputNetToCSV(net)
net=pg.readNetwork()
pg.showNetByAllMode(net)
```

**Step7: connect POIs to network**

```python
net=og.getNetFromOSMFile(osm_filename='/content/osm_datasets/tempe_asu.osm',network_type=('auto'),POIs=True)
og.connectPOIWithNet(net)
og.outputNetToCSV(net)
net=pg.readNetwork()
pg.showNetByAllMode(net)
```
